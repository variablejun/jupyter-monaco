```python
import pandas as pd
import numpy as np
import random
import string
```

```python
def score():
    return np.random.randint(0, 100)
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
df = pd.DataFrame([[score() for i in range(4)] for i in range(10)], index=[id() for i in range(10)], columns=["국어","수학","영어","사회"])
```

```python
# 과학추가 , 기존 학생들에게 랜덤 추가
```
